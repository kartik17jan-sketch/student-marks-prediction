```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# to visualise the results
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#loading data 
Data = "http://bit.ly/w-data"
raw_data = pd.read_csv(Data)

raw_data.head()
```

```python
#checking dataframe info
raw_data.info()
```

```python
#plotting the data
sns.set_style("dark")
plt.figure(figsize=(14,7))#setting the size of figure
plt.scatter(raw_data['Hours'],raw_data['Scores']) #plotting datapoints
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

```python
raw_data['Hours']
```

```python
raw_data['Scores']
```

```python
#converting the data into numpy array 

#X variables corresponds to the hours 
x = raw_data[["Hours"]].values   # values lib converted the data into numpy array #dobule brackets are used to maintain the structure

#Y variables corresponds to the scores
y = raw_data[["Scores"]].values
```

```python
x
```

```python
y
```

```python
print(x.shape)
print(y.shape)
```

```python
#splitting the data into train and test set
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0) 
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#define the model 
model = LinearRegression()

#fit the model
model.fit(x_train,y_train)
```

```python
# After using the model we would likely to see the best fit line 
plt.figure(figsize=(14,7))
sns.regplot(x = x_train, y = y_train)
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.title("Regression Line")
plt.show()
```

```python
#Model Evaluation
#we are evaluating the model on unseen data
y_pred = model.predict(x_test)
y_pred
```

```python
#calculating the error between the predicted outcomes and y_test
#we can also use mean squared error as we have studied in the notes
MAE = mean_absolute_error(y_test, y_pred)
MAE
```
